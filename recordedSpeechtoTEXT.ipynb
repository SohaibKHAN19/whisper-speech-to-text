```python
import gradio as gr
import transformers
from transformers import pipeline
```

```python
p = pipeline("automatic-speech-recognition",model ="openai/whisper-tiny")
```

```python
#ui
def transcribe(audio, state=""):
    if audio is None:
        return state, state
    try:
    # Your transcription logic
        result = p(audio,return_timestamps=True)["text"]
        new_state = state + result + " "
        return new_state, new_state
    # Update the state (adding new text to the old text)
    except Exception as e:
        return f"Error: {str(e)}", state
```

```python
gr.Interface(
    fn=transcribe,
    inputs=[
        gr.Audio(sources=["microphone","upload"], type='filepath'),
        gr.State("") # This is the hidden 'state' input
    ],
    outputs=[
        gr.Textbox(label="Transcript"), # First 'state' goes here
        gr.State()                       # Second 'state' stays hidden here
    ],
    live=True
).launch()
```
